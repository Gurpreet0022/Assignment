### Data Wrangling & Statistical Analysis 

```python
import os
import pandas as pd
from supabase import create_client
from pymongo import MongoClient
from dotenv import load_dotenv

# 1. Load credentials 
load_dotenv()

def fetch_data():
    # --- Supabase Setup ---
    SUPABASE_URL = os.getenv("SUPABASE_URL")
    SUPABASE_KEY = os.getenv("SUPABASE_KEY")
    supabase = create_client(SUPABASE_URL, SUPABASE_KEY)
    
    # Fetch from Supabase 
    sb_response = supabase.table("customers").select("*").execute()
    df_supabase = pd.DataFrame(sb_response.data)
    print(f"Fetched {len(df_supabase)} rows from Supabase.")

    # --- MongoDB Setup ---
    MONGO_URI = os.getenv("MONGO_URI")
    mongo_client = MongoClient(MONGO_URI)
    db = mongo_client["test"]
    collection = db["final_activity_logs"]
    
    # Fetch from MongoDB
    mongo_data = list(collection.find({}, {"_id": 0})) 
    df_mongodb = pd.DataFrame(mongo_data)
    print(f"Fetched {len(df_mongodb)} documents from MongoDB.")

    return df_supabase, df_mongodb

if __name__ == "__main__":
    df_sb, df_mg = fetch_data()
```

```python
# Load Dataset
df = pd.merge(df_sb, df_mg, on="customer_id", how="inner")
print(df.head())
```
